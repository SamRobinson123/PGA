## EDA

```python
df = pd.read_csv("C:/GitHub/PGA/raw_data.csv")
```

## Players With Highest Strokes Gained Putting

```python
## Vizualize Players With the Best Strokes Gained Putting Between 2015-2022
putting_data = df[['player', 'sg_putt', 'tournament_id']].dropna()

# Count the number of events per player
event_counts = putting_data.groupby('player')['tournament_id'].nunique()

# Filter players who have played in at least 15 events
eligible_players = event_counts[event_counts >= 70].index

# Filter the putting data to include only eligible players
filtered_data = putting_data[putting_data['player'].isin(eligible_players)]

# Calculate the average sg_putt for each eligible player
average_sg_putt = filtered_data.groupby('player')['sg_putt'].mean().sort_values(ascending=False).head(10)

# Plotting the top 10 players with the highest average sg_putt
plt.figure(figsize=(12, 8))
plt.barh(average_sg_putt.index, average_sg_putt, color='skyblue')
plt.xlabel('Average SG: Putting')
plt.title('Top 10 Players with Highest Average SG: Putting (2015-2022)')
plt.gca().invert_yaxis()  # To display the highest value at the top
plt.show()
```

```python
## Vizualize Players With the Best Strokes Gained Tee To Green Between 2015-2022
t2g_data = df[['player', 'sg_t2g', 'tournament_id']].dropna()

# Count the number of events per player
event_counts = t2g_data.groupby('player')['tournament_id'].nunique()

# Filter players who have played in at least 70 events
eligible_players = event_counts[event_counts >= 70].index

# Filter the t2g data to include only eligible players
filtered_data = t2g_data[t2g_data['player'].isin(eligible_players)]

# Calculate the average sg_t2g for each eligible player
average_t2g = filtered_data.groupby('player')['sg_t2g'].mean().sort_values(ascending=False).head(10)

# Plotting the top 10 players with the highest average sg_t2g
plt.figure(figsize=(12, 8))
plt.barh(average_t2g.index, average_t2g, color='skyblue')
plt.xlabel('Average SG: Tee To Green')
plt.title('Top 10 Players with Highest Average SG: Tee To Green (2015-2022)')
plt.gca().invert_yaxis()  # To display the highest value at the top
plt.show()
```
